## 2. CET1 Ratio Comparison (Interactive Plotly Chart)

```python
import plotly.graph_objs as go
import plotly.io as pio
pio.templates.default = None

scenarios = ['Baseline','CRE Collapse','Supervisory','Reverse ST']
vals = [9.03, 4.8, 3.1, 4.3]

fig = go.Figure(data=[go.Bar(x=scenarios, y=vals)])
fig.add_trace(go.Scatter(x=scenarios, y=[4.5]*len(scenarios), mode='lines', name='CET1 Threshold (4.5%)', line=dict(color='red', dash='dash')))
fig.update_layout(title='CET1 Ratios Across Scenarios',plot_bgcolor='white',paper_bgcolor='white', yaxis_title='CET1 Ratio (%)')
fig.show()
```

## 4. Reverse Stress Testing Heatmap (Plotly)

This heatmap visualizes the impact of various combinations of Commercial Real Estate (CRE) Decline (%) and Default Rate Increase on the bank's financial stability, specifically related to losses that could push the CET1 ratio below its viability threshold.

**How to Interpret the Heatmap:**

X-axis (CRE Decline (%)): This represents the percentage decline in Commercial Real Estate valuations. A higher value indicates a more severe CRE market downturn.

Y-axis (Default Rate Increase): This represents the increase in corporate default rates. A higher value indicates more widespread defaults among borrowers.

Colorscale (Z value): The color of each point on the heatmap (derived from the 'Z' values in the code) indicates the total losses incurred under that specific combination of CRE decline and default rate increase. Typically, in such heatmaps, one side (e.g., warmer colors) would represent conditions that lead to breaching the viability threshold, while the other side (e.g., cooler colors) represents conditions where the bank remains solvent.

**The Viability Boundary:** The purpose of a Reverse Stress Test is to identify the 'breaking points' where the bank's CET1 ratio falls below a critical threshold (in this case, 4.5%). The heatmap helps to visually identify the combinations of CRE decline and default rate increase that push the bank into this critical zone. You'll observe a boundary where the colors shift, indicating the transition from acceptable conditions to scenarios where the bank's capital becomes inadequate.

For reference, the 'Reverse Stress Test (Viability Threshold)' scenario mentions that corporate defaults of +6.8 pp or a CRE decline of >42% were found to be sufficient to breach the CET1 threshold independently. The heatmap shows how these factors combine and interact.

```python
import numpy as np
import plotly.graph_objs as go

cre = np.linspace(0.2,0.5,30)
default = np.linspace(0.02,0.1,30)
Z = np.zeros((30,30))
for i,c in enumerate(cre):
    for j,d in enumerate(default): Z[j,i] = 12*c + 40*d

fig = go.Figure(data=go.Heatmap(z=Z, x=cre, y=default, colorscale='RdBu_r'))
fig.update_layout(title='Reverse Stress Heatmap — Viability Boundary',plot_bgcolor='white',paper_bgcolor='white',xaxis_title='CRE Decline (%)',yaxis_title='Default Rate Increase')
fig.show()
```
